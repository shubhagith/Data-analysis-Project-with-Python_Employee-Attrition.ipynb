```python
visuals = ['dept','numberOfProjects','timeSpent.company','workAccident','promotionInLast5years','salary']
visuals
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig = plt.subplots(figsize = (10,15))

for p,q in enumerate(visuals):    # p is for indexing of graphs and q is for having the values to the graphs
  plt.subplot(3,2,p+1)
  plt.subplots_adjust(hspace = 1.0)
  sns.countplot(x = q, data = data, hue='left')
  plt.xticks(rotation = 90)
  plt.subplot(3,2,p+1)
```
